```python
import pandas as pd
import os
import shutil
import time
```

```python
# 총 88개 폴더명 리스트 만들기
train_df = pd.read_csv("/content/drive/MyDrive/efficientnet/전인서/train_df.csv")
label_list = train_df["label"].unique().tolist()
label_list = sorted(label_list)
len(label_list)
```

```python
# 폴더 안 파일 개수를 가져오는 함수
def get_num_of_files(directory):
  dirListing = os.listdir(directory)
  return len(dirListing)

# 15개 good 폴더의 파일 개수만 출력해보기
# 최초에는 good 폴더의 파일 개수 최솟값이 60이었다.
for label in label_list:
  num_files = get_num_of_files(f'/content/drive/MyDrive/efficientnet/전인서/train/{label}/')
  if num_files >= 60:
    print(num_files)
```

### -> good 폴더들은 보통 파일 개수가 200개 대이므로 이상치 폴더들도 200개까지 오버샘플링해준다.

```python
# 이상치 폴더명 리스트
anomaly_label_list = []

for label in label_list:
  if label[-4:] != 'good': # good이 아닌 것만 선택
    anomaly_label_list.append(label)

len(anomaly_label_list) # 88-15
```

```python
# 이상치 폴더 73개에 대해 for loop
# 폴더 내 파일들을 복사하되, 폴더 내 파일 개수가 200개 이상이 되면 stop

for i in range(len(anomaly_label_list)):
  current_dir = f'/content/drive/MyDrive/efficientnet/전인서/train/{anomaly_label_list[i]}'
  
  # 최초의 이미지들에 대해 복사를 반복하기 위해 리스트를 따로 만들어 둔다.
  original_file_list = []
  for file in os.listdir(current_dir):
    if len(file)== 9: # 앞에 time이 붙지 않은 파일이어야 하므로 길이가 9인 파일명만 담는다. ex) '13731.png'
      original_file_list.append(file)
  
  # 폴더 내 파일명 전체를 리스트에 담는다.
  whole_file_list = os.listdir(current_dir)

  # 폴더 내 파일 개수가 200개 미만이면 최초의 이미지들에 대해 복사를 반복한다.
  while len(whole_file_list) < 200:
    for file in original_file_list:
      t = float(time.time())
      shutil.copyfile(f'{current_dir}/{file}', f'{current_dir}/{t}-{file}')
    whole_file_list = os.listdir(current_dir)
  
  print(f'{i+1} {anomaly_label_list[i]}: completed')
```

```python
# 데이터 증강을 완료했다면
# 88개 폴더들 중 파일 개수가 200개가 안 되는 폴더가 있는지 확인해보자.
# 아무것도 출력되지 않으면 모든 폴더의 파일 개수는 200개 이상이다.

for label in label_list:
  num_files = get_num_of_files(f'/content/drive/MyDrive/efficientnet/전인서/train/{label}/')
  if num_files < 200:
    print(label)
```
